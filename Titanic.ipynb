```python
#Pclass 1 their age will marked as 37
#Pclass 2 their age will marked as 29
#Pclass 1 their age will marked as 24
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cmap='viridis')
```

```python
def input_age(cols):
    Age = cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass ==1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
            return Age
```

```python
df['Age']=df[['Age', 'Pclass']].apply(input_age,axis =1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cmap='viridis')
```

```python
df.drop('Cabin', axis=1,inplace=True)
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cmap='viridis')
```

```python
df.drop(['Sex', 'Embarked', 'Name', 'Ticket', 'PassengerId','Pclass'], axis=1, inplace=True)
```

```python
df=pd.concat([df,sex,Embark], axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Survived', axis=1)
y = df['Survived']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
x_test
```

```python
prediction=logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)
```
